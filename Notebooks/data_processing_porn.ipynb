```python
# Instantiates a client
client = speech.SpeechClient()
```

```python
def create_output(client, video_type, set_n, label):

    audio_file_type = 'mp3'
    vid_output = f'{video_type}/set{set_n}'
    time_per_segment = 30
    cutoff = 10
    segment_folder = f'{video_type}_segments/set{set_n}'
    seg_list = generate_segment_list(label, audio_file_type, vid_output,
                                     client, time_per_segment, cutoff, segment_folder)
    
    with open(f'output/{video_type}/{video_type}_set{set_n}.pkl', 'wb') as filename:
        pickle.dump(seg_list, filename)
        filename.close()
        
    seg_df = create_df(seg_list)
    seg_df.to_pickle(f'output/{video_type}/{video_type}{set_n}_df.pkl')
    
    del seg_list
    gc.collect()
    
    return f'Finished with {video_type}{set_n}\n\n\n'
```

```python
for idx in range(1, 6):
    create_output(client, 'ph', idx, 1)
```
